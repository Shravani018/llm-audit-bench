#### 02: Transparency Score

**Loading model's card text and scoring it aginst 7 completeness criteria.**

```python
import sys
import os
import yaml
from dataclasses import asdict
import json
sys.path.append(os.path.abspath(".."))
# from src import extracting_metadata
from src import transparency_scoring 
import warnings
warnings.filterwarnings('ignore')
```

```python
with open("../config.yaml", "r") as f:
    config = yaml.safe_load(f)
models=config['models']['list']
```

```python
# Defining the criteris and weights(summing to 1.0)
criteria={
    "has_model_card":0.20,
    "license":0.15,
    "training_data":0.20,
    "limitations":0.15,
    "intended_use":0.10,
    "evaluation_results":0.10,
    "carbon_footprint":0.10
}
```

```python
# Evaluating transperancy for each model
model_transperancy=[transparency_scoring.eval_transperancy(model_id) for model_id in models]
```

```python
rows = []
for r in model_transperancy:
    row = {"model_id": r["model_id"], "transparency_score": r["score"]}
    row.update(r["checks"])  # flatten checks into columns
    rows.append(row)
```

```python
os.makedirs("../results",exist_ok=True)
with open("../results/transparency_scores.json","w") as f:
    json.dump(rows, f, indent=2)
```

**Conclusion**
- Transparency scores range from 0.55 to 1.0, showing that documentation quality varies across models.
- Most models include basic information like license and training data, but sections such as limitations, intended use, and evaluation results are often missing.
- `TinyLlama` scores lowest due to missing key documentation fields, while models like `phi-1_5` and `stablelm` are more complete.
- Carbon footprint reporting is mostly absent, with only one model including it.
- Transparency mainly reflects how well a model is documented, not how it actually behaves.
- High transparency scores are useful, but do not guarantee reliability, safety, or fairness.
